Import Libraries

```python
import numpy as np
import pandas as pd
```

Load DataSet

```python
df=pd.read_csv('drug200.csv')
df
```

Declare X & Y

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
x,y
```

Filling Missing Values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'most_frequent')

imputer = imputer.fit(x[:,0:4])
x[:,0:4]= imputer.transform(x[:, 0:4])
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder_X = LabelEncoder()
#CHANGE THE INDEX FOR APPROPRIATE COLUMNS OF X
x[:,1] = labelEncoder_X.fit_transform(x[:,1])
x[:,2] = labelEncoder_X.fit_transform(x[:,2])
x[:,3] = labelEncoder_X.fit_transform(x[:,3])

labelEncoder_y = LabelEncoder()
y = labelEncoder_y.fit_transform(y)
y
```

X Inputs After Label Encoding

```python
x
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

Apply Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(x_train, y_train)
```

Y-Pred for DT

```python
y_pred_dt = decision_tree.predict(x_test)
y_pred_dt
```

Evaluation (Accuracy Score, Classification Report , Confusion Matrix)

```python
from sklearn.metrics import accuracy_score, classification_report,  confusion_matrix
accuracy_dt = accuracy_score(y_test, y_pred_dt)
report_dt = classification_report(y_test, y_pred_dt)
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_dt))
```

Print Accuracy & Classification Report

```python
print("Decision Tree Classifier:")
print("Accuracy:", accuracy_dt)
print("Classification Report:")
print(report_dt)
```

HyperParameter Tuning

```python
print("GG0",format(decision_tree.score(x_train,y_train)))
print("FF",format(decision_tree.score(x_test,y_test)))
```

```python
from sklearn.model_selection import GridSearchCV
dt = {
    'max_depth': [3, 5, 7, 10, None]
}
grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), dt, cv=5)
grid_dt.fit(x_train, y_train)

print("Best parameters for Decision Tree:", grid_dt.best_params_)
```

Apply Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(x_train, y_train)
```

Y-Predist for RF

```python
y_pred_rf = random_forest.predict(x_test)
y_pred_rf
```

Evaluation (Accurcy, Classification Report, Confusion Matrix)

```python
accuracy_rf = accuracy_score(y_test, y_pred_rf)
report_rf = classification_report(y_test, y_pred_rf)
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
```

Print Accuracy & Classification Report for RF

```python
print("Random Forest Classifier:")
print("Accuracy:", accuracy_rf)
print("Classification Report For Random Forest:")
print(report_rf)
```

HyperParameter Tuning for RF

```python
rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7, 10, None]
}
grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), rf, cv=5)
grid_rf.fit(x_train, y_train)

print("Best parameters for Random Forest:", grid_rf.best_params_)
```
