```python
import numpy as np 
import pandas as pd 
```

```python
df=pd.read_csv('Clean_travel.csv')
```

```python
df.head()
```

```python
x=df.drop('ProdTaken',axis=1)
x
```

```python
y=df['ProdTaken']
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

## ONE HOT ENCODING

```python
cat_features = x.select_dtypes(include="object").columns
num_features = x.select_dtypes(exclude="object").columns

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

numeric_transformer = StandardScaler()
oh_transformer = OneHotEncoder(drop='first')

preprocessor = ColumnTransformer(
    [
         ("OneHotEncoder", oh_transformer, cat_features),
          ("StandardScaler", numeric_transformer, num_features)
    ]
)

preprocessor


x_train=preprocessor.fit_transform(x_train)


pd.DataFrame(x_train)

x_test=preprocessor.transform(x_test)
```

## Model Training & Testing

```python
from sklearn.ensemble import AdaBoostClassifier

classifier=AdaBoostClassifier()

classifier.fit(x_train,y_train)
```

```python
x_train_pred=classifier.predict(x_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
matrix=confusion_matrix(y_train,x_train_pred)
report=classification_report(y_train,x_train_pred)
acc=accuracy_score(y_train,x_train_pred)

print('For Training Set')
print("Accuracy Score: ",acc)
print('Confusion Matrix:')
print(matrix)
print("Classification Report: ")
print(report)

print("-----------------------------")
print("For Test Set")
matrix=confusion_matrix(y_test,y_pred)
report=classification_report(y_test,y_pred)
acc=accuracy_score(y_test,y_pred)

print("Accuracy Score: ",acc)
print('Confusion Matrix:')
print(matrix)
print("Classification Report: ")
print(report)
```

### HyperParameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={
    'n_estimators':[50,60,70,80,90],
    'algorithm':['SAMME','SAMME.R']
}

params
```

```python
grid=GridSearchCV(estimator=classifier,param_grid=params,cv=3,verbose=2,n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_index_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(x_test)
```

```python
x_train_pred=grid.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
matrix=confusion_matrix(y_train,x_train_pred)
report=classification_report(y_train,x_train_pred)
acc=accuracy_score(y_train,x_train_pred)

print('For Training Set For HyperParameter Tuning')
print("Accuracy Score: ",acc)
print('Confusion Matrix:')
print(matrix)
print("Classification Report: ")
print(report)

print("-------------------------------------------")
print("For Testing Set For HyperParameter Tuning")
matrix=confusion_matrix(y_test,y_pred)
report=classification_report(y_test,y_pred)
acc=accuracy_score(y_test,y_pred)

print("Accuracy Score: ",acc)
print('Confusion Matrix:')
print(matrix)
print("Classification Report: ")
print(report)
```
