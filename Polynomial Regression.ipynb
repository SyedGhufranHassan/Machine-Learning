```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#apply numpy to find x & y 

x= 6 * np.random.rand(100, 1) - 3
y=0.5 * x**2 + 1.5*x + 2 + np.random.rand(100,1)

x,y
```

```python
#scatter plot
plt.scatter(x,y,color='black')
plt.xlabel('X')
plt.ylabel("Y")
```

```python
#train-test-split
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#implement simple liear regression
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#perfomance metrics of simple linear regression
from sklearn.metrics import r2_score 

r2=r2_score(y_test,regressor.predict(x_test))
r2
```

```python
#check the best for line
plt.plot(x_train,regressor.predict(x_train))
plt.scatter(x_train,y_train)
```

```python
#lets apply polynomial regression with degree==2
from sklearn.preprocessing import PolynomialFeatures
polynomial=PolynomialFeatures(degree=2,include_bias=True)

x_train_poly=polynomial.fit_transform(x_train)
x_test_poly=polynomial.transform(x_test)
```

```python
x_train_poly
```

```python
x_test_poly
```

```python
#r2 score for polynomial regression for degree=2
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(x_train_poly,y_train)

y_pred=regression.predict(x_test_poly)

r2s=r2_score(y_test,y_pred)
r2s
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
plt.scatter(x_train,regression.predict(x_train_poly))
plt.scatter(x_train,y_train)
```

```python
#lets apply polynomial regression with degree==3
from sklearn.preprocessing import PolynomialFeatures
polynomial1=PolynomialFeatures(degree=3,include_bias=True)

x_train_poly1=polynomial1.fit_transform(x_train)
x_test_poly1=polynomial1.transform(x_test)
```

```python
x_train_poly1
```

```python
#r2 score for polynomial regression for degree=3
from sklearn.metrics import r2_score
regression1=LinearRegression()
regression1.fit(x_train_poly1,y_train)

y_pred=regression1.predict(x_test_poly1)

r2sc=r2_score(y_test,y_pred)
r2sc
```

```python
#prediction for new data
x_new=np.linspace(-3,3,200).reshape(200,1)
x_new_poly=polynomial1.transform(x_new)
```

```python
y_new=regression1.predict(x_new_poly)
plt.plot(x_new,y_new,"-r",linewidth=2, label="New Predictions")
plt.plot(x_train,y_train,"b.",label="Training Pints")
plt.plot(x_test,y_test,"g.",label="Testing points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

PipeLine

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    X_new=np.linspace(-3,3,200).reshape(200,1)
    
    poly_features=PolynomialFeatures(degree=degree,include_bias=True)
    line_reg=LinearRegression()
    polyno_regression=Pipeline([
        ("poly_features",poly_features),
        ("line_reg",line_reg)
    ])
    #polynomial and linear regression 
    polyno_regression.fit(x_train,y_train)
    y_pred_new=polyno_regression.predict(X_new)
    
    #plotting prediction line
    plt.plot(X_new,y_pred_new,"-r", label="Degree" + str(degree), linewidth=2)
    plt.plot(x_train,y_train,"b.",linewidth=3)
    plt.plot(x_test,y_test,"g.",linewidth=3)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis([-4,4,0,10])
    plt.legend(loc="upper left")
    plt.show()
```

```python
poly_regression(5)
```
