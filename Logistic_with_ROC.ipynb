```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_classes=2,random_state=42)
```

```python
x,y
```

```python
#Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#import model & roc curve & roc accuracu score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from matplotlib import pyplot
```

```python
#create a dummy model with default 0 as output
dummy_model_prob=[0 for i in range(len(y_test))]
dummy_model_prob
```

```python
#lets create basic logistic regression model
model=LogisticRegression()

model.fit(x_train,y_train)
```

```python
#prediction based on probability
model_prob=model.predict_proba(x_test)
model_prob
```

```python
#lets focus on positive outcome
model_prob=model_prob[:,1]
```

```python
#lets calculate the scores
dummy_model_auc=roc_auc_score(y_test,dummy_model_prob)
model_auc=roc_auc_score(y_test,model_prob)
print("DUMMY MODEL: ",dummy_model_auc)
print("REAL MODEL: ",model_auc)
```

```python
#calculate ROC curves
dummy_fpr,dummy_tpr, _ =roc_curve(y_test,dummy_model_prob)
model_fpr,model_tpr,thresholds =roc_curve(y_test,model_prob)
```

```python
thresholds
```

```python
#plot the ROC curve for the model
pyplot.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy Model')
pyplot.plot(model_fpr,model_tpr,marker='.',label='Logistic Model')

pyplot.xlabel('False Positive Rete')
pyplot.ylabel('True Positive Rate')

pyplot.legend()    
pyplot.show()
```

```python
#plot the ROC curve for the mdoel
fig=pyplot.figure(figsize=(20,50))

ax=fig.add_subplot(111)
for xyz in zip(model_fpr,model_tpr,thresholds):
    ax.annotate('%s' % np.round(xyz[2],2), xy=(xyz[0],xyz[1]))
pyplot.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy Model')
pyplot.plot(model_fpr,model_tpr,marker='.',label='Logistic Model')
pyplot.xlabel('False Positive Rete')
pyplot.ylabel('True Positive Rate')

pyplot.legend()    
pyplot.show()
```

From this Graph now we can select thresholds where e can find maximize True positive rate and minimum False positive rate.
