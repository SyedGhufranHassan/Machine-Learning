```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import make_classification
from sklearn.datasets import make_classification
```

```python
#create dataset
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
#independent features
pd.DataFrame(x)
```

```python
#dependent features
y
```

```python
#train-test-split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#implement Logistic Regression
from sklearn.linear_model import LogisticRegression

logistic=LogisticRegression()
```

```python
#apply logistic on x_train,y_train
logistic.fit(x_train,y_train)
```

```python
#apply prediction on x_test
y_pred=logistic.predict(x_test)
```

```python
#predicted y
y_pred
```

```python
#probability of x_test data points of belongs to classes
logistic.predict_proba(x_test)
```

```python
#perfomance metrics
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,classification_report


```

```python
#Confusion matrix
confusion=confusion_matrix(y_test,y_pred)
print("CONFUSION MATRIX for LOGISTIC REGRESION: ")
print(confusion)
```

```python
#accuracy
accu=accuracy_score(y_test,y_pred)
print("ACCURACY for LOGISTIC REGRESION: ",accu)
```

```python
#precision score
pre=precision_score(y_test,y_pred)
print("PRECISION SCORE for LOGISTIC REGRESION: ",pre)
```

```python
#recall score
rec=recall_score(y_test,y_pred)
print("RECALL SCORE for LOGISTIC REGRESION: ",rec)
```

```python
#Classification report
report=classification_report(y_test,y_pred)
print("CLASSIFICATION REPORT for LOGISTIC REGRESION: ",report)

```

HYPER PARAMETER TUNING & CROSS VALIDATION

GRID SEARCH CV

```python
model=LogisticRegression()
penalty=["l1", "l2", "elasticnet"]
c_values=[100,10,1.0,0.1,0.01]
solver=["lbfgs", "liblinear", "newton-cg", "newton-cholesky", "sag","saga"]
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
#GridSearch CV
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred1=grid.predict(x_test)
```

```python
report=classification_report(y_test,y_pred1)
print("CLASSIFICATION REPORT for GRIDSEARCHCV: ",report)

accu=accuracy_score(y_test,y_pred1)
print("ACCURACY for GRIDSEARCHCV: ",accu)

confusion=confusion_matrix(y_test,y_pred1)
print("CONFUSION MATRIX for GRIDSEARCHCV: ")
print(confusion)
```

RANDOMIZED SEARCH CV:

```python
from sklearn.model_selection import RandomizedSearchCV

model=LogisticRegression()
random=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
random.fit(x_train,y_train)
```

```python
y_pred2=random.predict(x_test)
```

```python
report=classification_report(y_test,y_pred2)
print("CLASSIFICATION REPORT for RandomizedSearchCV: ",report)

accu=accuracy_score(y_test,y_pred2)
print("ACCURACY for RandomizedSearchCV: ",accu)

confusion=confusion_matrix(y_test,y_pred2)
print("CONFUSION MATRIX for RandomizedSearchCV: ")
print(confusion)
```

LOGISTIC REGRESSION FOR IMBALANCED DATASET:

```python
from collections import Counter
from sklearn.datasets import make_classification
```

```python
#create dataset
x1,y1= make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,n_redundant=0,weights=[0.99],random_state=10)
```

```python
#imabalnced dataset
y1
```

```python
Counter(y1)
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver,class_weight=class_weight)
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

cv=StratifiedKFold()
grid1=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv)
```

```python
grid1.fit(x1_train,y1_train)
```

```python
grid1.best_params_
```

```python
y_pred=grid1.predict(x1_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix


acc=accuracy_score(y1_test,y_pred)
report=classification_report(y1_test,y_pred)
matrix=confusion_matrix(y1_test,y_pred)

print("FOR IMBALANCED DATASET:")
print("ACCURACY: ",acc)
print("classification_report: ")
print(report)
print("CONFUSION MATRIX: ")
print(matrix)
```
