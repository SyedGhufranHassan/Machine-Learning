```python
from sklearn.model_selection import train_test_split
```

```python
#now it gives us independent and dependent feature.
x,y=load_iris(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
acc=accuracy_score(y_test,y_pred)
report=classification_report(y_test,y_pred)
```

```python

print("Accuracy Score for Guassian NB: ",acc)
print("Classification Report for Guassian NB: ")
print(report)
```

Now we import Tips data from Seabron and apply # different varients of Naive Bayes.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.naive_bayes import GaussianNB
```

```python

# Load the dataset
df = sns.load_dataset('tips')

# Define features and target
x = df[['total_bill', 'tip', 'sex', 'smoker', 'day', 'size']]
y = df['time']
x,y
```

```python
# Split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Initialize LabelEncoders for target variable (y)
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
# Define ColumnTransformer for encoding categorical columns
encoder = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(drop='first'), ['sex', 'smoker', 'day'])  # OneHotEncode all categorical columns
    ],
    remainder='passthrough'  # Keep the other columns ('total_bill', 'tip', 'size') as they are
)
```

```python
# Fit and transform the training data
x_train = encoder.fit_transform(x_train)

# Only transform the test data (using the same encoder fit on training data)
x_test = encoder.transform(x_test)
```

Gaussian NB

```python
# Initialize and fit Gaussian Naive Bayes model
gnb = GaussianNB()
gnb.fit(x_train, y_train)

# Predict using the test data
y_pred = gnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
acc=accuracy_score(y_test,y_pred)
report=classification_report(y_test,y_pred)

print("Accuracy Score for Guassian NB For tips data: ",acc)
print("Classification Report for Guassian NB for tips data: ")
print(report)
```

MultiNomial NB:

```python
from sklearn.naive_bayes import MultinomialNB
mn=MultinomialNB()
mn.fit(x_train,y_train)

y_pred=mn.predict(x_test)


from sklearn.metrics import accuracy_score,classification_report
acc=accuracy_score(y_test,y_pred)
report=classification_report(y_test,y_pred)

print("Accuracy Score for MultiNomial NB Fir Tips data: ",acc)
print("Classification Report for MultiNomial NB for Tips data: ")
print(report)
```

Bernouli NB

```python
from sklearn.naive_bayes import BernoulliNB
ber=BernoulliNB()
ber.fit(x_train,y_train)

y_pred=ber.predict(x_test)


from sklearn.metrics import accuracy_score,classification_report
acc=accuracy_score(y_test,y_pred)
report=classification_report(y_test,y_pred)

print("Accuracy Score for Bernouli NB Fir Tips data: ",acc)
print("Classification Report for Bernouli NB for Tips data: ")
print(report)
```
